```python
data['buying'] = data['buying'].replace('düsük','low')
data['doors'] = data['doors'].replace(['5+','iki','44'],['5more','2','4'])#44 ün 4 olabileceğini düşünüp ona göre işlem yaptım.
```

```python
data = data.dropna()
#neww_data = data[data['buying'] != '?']
#neww_data = data[data['buying'] != '-']
#neww_data = data[~data['buying'].isin(['?', '-','*'])]
```

```python
new_data = pd.get_dummies(data, columns = data.columns,
                             prefix=data.columns)
```

```python
y_data = data['class'].replace(['unacc','acc','good','vgood'],['-1','0','1','2'])
```

Model kurulumu

```python
from sklearn.model_selection import train_test_split ,cross_val_score,RandomizedSearchCV
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#new_data = new_data.drop(['class_unacc','class_acc','class_good','class_vgood'],axis=1)
```

dt model vanilla



```python
X_train , x_test , y_train ,y_test = train_test_split(new_data,y_data,test_size=0.2,random_state=13)
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
```

dt model tuned

```python
dt_model_tuned = DecisionTreeClassifier()
param_dist = {
    'criterion':['gini','entropy'],
    'max_depth': np.arange(2,13,1),
    'min_samples_leaf' : [1, 2, 4],
    'max_features' : ['auto', 'sqrt']
}
random_dt_model = RandomizedSearchCV(dt_model_tuned,param_dist,n_iter=35,cv=5)
random_dt_model.fit(X_train,y_train)
```

dt model rastgele +tune edilecek

```python
dt_model_2 = DecisionTreeClassifier(
                criterion='entropy',
                splitter='best',
                max_depth=7,
                min_samples_split=3,
                max_features=21,)
dt_model_2.fit(X_train,y_train)
```

rf_model vanilla

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train,y_train)
```

```python
cross_val_score(rf_model,new_data,y_data,cv=7).mean()
```

```python
cross_val_score(dt_model_2,new_data,y_data,cv=7).mean()
```

```python
cross_val_score(dt_model,new_data,y_data,cv=7).mean()
```

```python
y_pred = dt_model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
y_pred_1 = rf_model.predict(x_test)
print(classification_report(y_test,y_pred_1))
```
